```python
daniel_part_1 = [158938, 84487, 156658, 169374, 162896, 257863, 620501, 778148, 728680] 
daniel_part_2 = [159351, 84887, 195490, 190022, 190003, 271312, 7157320016, 153987400289, 30921618956]
daniel_label = "Intel i7-8565U (8) @ 1.80GHz, 16GB memory"

simon_part_1 = [273750, 100995, 279417, 327998, 272807, 773752, 1020436, 3362821, 1265888]
simon_part_2 = [269255, 97596, 264143, 266028, 270304, 794011, 13306030687, 50094187973, 22552110883]
simon_label = "Intel i9-9880H (16) @ 2.30GHz, 16GB memory"

title_1 = "Part 1 benchmark"
title_2 = "Part 2 benchmark"

NAMES = []
NAME = "circuit"
for i in range(1,10):
    NAMES.append(NAME+str(i))

units = "[ns/op]"
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.gcf().subplots_adjust(bottom=0.15)
```

```python
def draw_bars(bars_1, label_1, bars_2, label_2, title):
    x = np.arange(len(NAMES))
    
    fig = plt.figure()
    
    ax = fig.add_axes([0,0,1,1])
    ax.set_yscale('log')
    ax.bar(x - 0.15, bars_1, color = 'b', width = 0.3, label=label_1)
    ax.bar(x + 0.15, bars_2, color = 'r', width = 0.3, label=label_2)
    ax.legend()
    
    plt.title(title)
    plt.xticks(x, NAMES)
    plt.ylabel(units, rotation="horizontal")
    plt.grid()
    plt.savefig(title.replace(" ", "_")+".png", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
draw_bars(daniel_part_1, daniel_label, simon_part_1, simon_label, title_1)
```

```python
draw_bars(daniel_part_2, daniel_label, simon_part_2, simon_label, title_2)
```
